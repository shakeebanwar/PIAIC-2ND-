```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],'data1': range(7)})
df2 = pd.DataFrame({'key': ['a', 'b', 'd'],'data2': range(3)})
print(df1)
df2
```

```python
pd.merge(df1,df2) # jo dono ma honga unko merge karaga 
```

```python
pd.merge(df1,df2,on = 'key')
#agr simple 2 dataframe ko merge karaga tu 2no ma 1 column same hona chaye yaha humna on='key ' diya means iski base pa 
#merge karna by default bi inhi ki base pa hota 
```

```python
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],'data1': range(7)})
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'],'data2': range(3)})
print(df3)
df4
```

```python
pd.merge(df3,df4,left_on = 'lkey', right_on = 'rkey')
# key same nhi hogi tu error generate kardaga
#inner same val ko merge karaga
#outer alag vali ko karaga jaha val nhi wah nan
```

```python
left = pd.DataFrame({'key1': ['foo', 'foo', 'bar'],
'key2': ['one', 'two', 'one'],
'lval': [1, 2, 3]})
right = pd.DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'],
'key2': ['one', 'one', 'one', 'two'],
'rval': [4, 5, 6, 7]})
print(left)
right
```

```python
pd.merge(left, right, on  = ['key1','key2'], how='outer') #ak sa zdda column ja basis pa merge
```

```python
pd.merge(left, right, on='key1')
```

```python
pd.merge(left, right, on='key1', suffixes = ('_left','_right')) # suffix key2-left ya iski waja
```
