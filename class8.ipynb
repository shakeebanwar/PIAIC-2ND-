# Discretization and Binning


```python
ages = [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]
```

```python
bins = [18, 25, 35, 60, 100]
# ya category wise karaga mean 18 sa 25 wala ak  25 sa 35 waali ak
```

```python
cats = pd.cut(ages, bins)
```

```python
cats
```

```python
cats.codes # ya category mean 18,25 ka index 0 ha tu ya us hisab sa index da raha ha 
```

```python
cats.categories # ya sari category ha jis ma divide hoa ha 
```

```python
pd.value_counts(cats)
 # ya bata raha ha kn sa index kitne baar ha 
```

```python
import pandas as pd
test = pd.DataFrame({'days': [0,31,45]})
test['range'] = pd.cut(test.days, [0,30,60])
test
```

```python
t=pd.DataFrame({'data1':[10,34,54,56]})
t['data2']=pd.cut(t.data1,[10,30,35,60])
t
# agar hum data1 ka column ki value ko compare kar raha ha tu 10 sa 30 ki ak category ha aur data1 ka column ki phali value
# 1 ha tu nhi nan aye ak agha sa compare hoga
```

```python
c=pd.cut(ages, [18, 26, 36, 61, 100], right=False)
c
#
```

```python
c.categories
```

```python
cats.categories
```

```python
group_names = ['Youth', 'YoungAdult', 'MiddleAged', 'Senior']
pd.cut(ages, bins, labels=group_names)
# haar category ko name diya ha simple
```

# Detecting and Filtering Outliers


```python
data = pd.DataFrame(np.random.randn(1000, 4))
data.describe()
```

```python
data
```

```python
col=data[2]
```

```python
col[np.abs(col)>2] #neg wali possitive ho raha ha jo grater than  2 ha
```

```python
data[(np.abs(data)>2)]
```

```python
data[(np.abs(data)>3)][22:25] # ya range ha simple
```

```python
data[(np.abs(data) > 3).any(1)] # wo row return kar raha ha koi row ma >3 ha tu usko return kar raha ha 
```

```python
np.sign(data) # haar value ka sath jo sign ha uska hisab sa return kar raha mean jaha 1 ha waha possitive -1 negative
```

```python
np.sign(data)*3 # haar value ka sign ka corresponding value
# haar value ka sath jo sign ha uska hisab sa return kar raha mean jaha 1 ha waha possitive -1 negative
```

```python
data[np.abs(data) > 3] = np.sign(data) * 3
data.describe()
```
